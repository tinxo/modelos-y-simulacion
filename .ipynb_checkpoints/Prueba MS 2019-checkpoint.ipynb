## Libreta Nro. 1 - Métodos de generación de números aleatorios

```python
# Importaciones iniciales
import numpy as np
import random as rn
```

```python
# Von Neumann
def generarVonNeumann(semilla, cantidadAGenerar):
    resultados = []
    resultados.append(semilla)
    while len(resultados) < cantidadAGenerar:
        semilla *= semilla
        txt = str(semilla)
        while len(txt) < 8:
            txt = '0' + txt
        semilla = int(txt[2:-2])
        if str(semilla)[-2:] == '00':
            txt = str(semilla)
            txt = txt[:-2] + '17'
            semilla = int(txt)
        resultados.append(semilla)
        print(semilla)
    return resultados
```

```python
print(generarVonNeumann(3245,290))
```
